Aumento o disminucion(Banda Ancha, Dial up) a lo largo de los años

A simple vista si analizamos el Dataset por encima, vemos que tiene 2 variables principales, la banda ancha fija y la dial up.
Tomando la banda ancha valores altos, mimentras que la dial up, tomo valores mucho mas chicos.
Por lo que veremos su comportamiento a lo largo de los anios

Cargo el Dataset, eliminamos valores nulos y vemos que tiene

```python
Ba = pd.read_csv('Datasets/Internet_BAF (1).csv')
Ba.dropna(inplace=True)
```

```python
Ba
```

Hacemos una manipulacion de datos, cambiando punto valor vacio y cambiando el tipo de dato, para poder graficar

```python
Ba['Banda ancha fija'] = Ba['Banda ancha fija'].str.replace('.','')
Ba['Dial up'] = Ba['Dial up'].str.replace('.','')
Ba['Total'] = Ba['Total'].str.replace('.','')

Ba.replace("- 0", 0,inplace=True)

Ba['Banda ancha fija'] = Ba['Banda ancha fija'].astype('int64')
Ba['Dial up'] = pd.to_numeric(Ba['Dial up'])
Ba['Total'] = Ba['Total'].astype('int64')
```

Al graficar podemos ver que en relacion al paso de los anios, la banda ancha fija va en aumento mientras que el dial up va disminuyendo

Eso se puede deber a que la banda ancha permite mayores velocidades y es una tecnologia mas moderna que la Dial up. Por lo que a medida que pasa el tiempo de a poco va reemplazando cada vez mas el Dial up, explicando el motivo por el cual este disminuye. Ya que la poblacion que usa Internet siempre va a buscar una mayor velocidad para tener mucha mas comodidad.

```python
sns.pairplot(Ba)
```

Graficando cada uno por separado podemos ver a parte de que la banda ancha si va en aumento y el dial up va disminuyendo, que lo valores de cada variable presentan una gran diferencia, mientras que en promedio la banda ancha esta cerca de los 300000 en el 2014, el dial up no supera los 2000, en ese mismo anio.

Concluyendo de que efectivamente el reemplazo y utilizacion de cada uno presentan tendencias opuestas

```python
sns.lineplot(data=Ba,x='Año',y='Banda ancha fija')
```

```python
sns.lineplot(data=Ba,x='Año',y='Dial up')
```

Aumento y disminucion porcentual de Acessos por tecnologia

Existe varias formas de transmitir a la hogares, todas presentan sus ventajas y desventajes. Pero vamos a ver, de esas tecnologias cuales son las que estan en aumento y cuales no.

Cargamos el dataset y eliminamos valores nulos

```python
At = pd.read_csv('Datasets/Internet_Accesos-por-tecnologia (1).csv')
At.dropna(inplace=True)
```

```python
At.info()
```

Hacemos una manipulacion de datos, cambiando punto valor vacio y cambiando el tipo de dato, para poder graficar

```python
At['ADSL'] = At['ADSL'].str.replace('.','')
At['Cablemodem'] = At['Cablemodem'].str.replace('.','')
At['Fibra óptica'] = At['Fibra óptica'].str.replace('.','')
At['Wireless'] = At['Wireless'].str.replace('.','')
At['Total'] = At['Total'].str.replace('.','')
At['Año'] = At['Año'].str.replace(' *','')
At['Año'] = At['Año'].str.replace('*','')

At.replace("- 0", 0,inplace=True)

At['ADSL'] = pd.to_numeric(At['ADSL'])
At['Cablemodem'] = pd.to_numeric(At['Cablemodem'])
At['Fibra óptica'] = At['Fibra óptica'].astype('int64')
At['Wireless'] = At['Wireless'].astype('int64')
At['Total'] = At['Total'].astype('int64')
At['Año'] = At['Año'].astype('int64')
```

En el grafico podemos ver como cada tipo de tecnologia va en aumento, menos ADSL(Internet por telefonia). Pero para poder apreciarlo mejor vamos a graficar cada una por separado, y sacar la variacion porcentul del promedio

```python
sns.pairplot(At)
```

Efectivamente vemos que ADSL va dismunuyendo, debido a que es una tecnologia que solo permite uso de bajas velocidades. Por lo que en una era del internet como la actual presenta una limitacion en el uso de la misma.

```python
sns.lineplot(data=At,x='Año',y='ADSL')
```

Para poder apreciar cual tecnologia es la que esta en auge y cual esta desapareciendo, vamos a sacar un porcentaje relativo del crecimiento, desde 2014 hasta 2021. Con el valor promedio de cada uno. No vamos a obtener un valor exacto, pero nos permitira ver el comportamiento y la tendencia de cada uno.

```python
past = At[At['Año'] == 2014]
present = At[At['Año'] == 2021]
```

```python
Padsl = ((present.ADSL.mean()-past.ADSL.mean())/past.ADSL.mean())*100
```

En los siguientes graficos podemos ver como la Fibra Optica, el Wireless, el Cablemodem, van en constante aumento debido a que es una tecnologia que permite mayores velocidades de Mbps(internet), dando paso a mayores avances en tecnologias de telecomunicacion, con la Fibra optica como la tacnologia que mas crecio en estos ultimos anios

```python
sns.lineplot(data=At,x='Año',y='Cablemodem')
```

```python
Pcable = ((present.Cablemodem.mean()-past.Cablemodem.mean())/past.Cablemodem.mean())*100
```

```python
sns.lineplot(data=At,x='Año',y='Wireless')
```

```python
Pwireless = ((present.Wireless.mean()-past.Wireless.mean())/past.Wireless.mean())*100
```

```python
sns.lineplot(data=At,x='Año',y='Otros')
```

```python
Potros = ((present.Otros.mean()-past.Otros.mean())/past.Otros.mean())*100
```

```python
sns.lineplot(data=At,x='Año',y='Fibra óptica')
```

```python
Pfibra = ((present['Fibra óptica'].mean()-past['Fibra óptica'].mean())/past['Fibra óptica'].mean())*100
```

Creo un Dataframe con todos los valores porcentuales de crecimiento, y los exporto a un csv

```python
Porcentajes = pd.DataFrame({'Tecnologia' : ['ADSL','Cablemodem', 'Fibra Optica', 'Wireless', 'Otros'],
                            'Porcentaje de Crecimiento' : [int(Padsl), int(Pcable), int(Pfibra), int(Pwireless), int(Potros)]})
```

Aca podemos ver de forma mas sencilla como es la tendencia de cada tecnologia. Como vemos la tecnologia ADSL y otros vienen en declive. Mientras que Cablemodem, Fibra Optica y Wireless van en aumento, siendo la Fibra Optica la tecnologia que mas avanzo estos anios. Todo se debe a que la Fibra optica es la que mayor velocidad permite de todas, si bien puede llegar a tener la infraestructura mas costosa, es compensado por la velocidad.

```python
Porcentajes
```

```python
Porcentajes.to_csv('Datasets/Porcentajes.csv', index=False)
```

Costo por Acceso

Un buen analisis seria poder ver cuanto a evolucionado el coste del acceso al ciudadano en promedios generales, y si fue en creciendo al aumentar y proveer mas velocidad o si disminuyo su costo

Cargo los datasets que se utilizaran y observo que hay en cada uno

```python
Iba = pd.read_csv('Datasets/Internet_BAF.csv')
Ii = pd.read_csv('Datasets/Internet_Ingresos.csv')
```

```python
Iba.head()
```

```python
Ii.head()
```

Hago una union de las dos tablas.
Hacemos una manipulacion de datos, cambiando punto por valor vacio y cambiando el tipo de dato, para poder graficar

```python
p = pd.merge(Iba,Ii,how='inner')
```

```python
p['Total'] = p['Total'].str.replace('.','')
p['Ingresos (miles de pesos)'] = p['Ingresos (miles de pesos)'].str.replace('.','')

p['Total'] = pd.to_numeric(p['Total'])
p['Ingresos (miles de pesos)'] = pd.to_numeric(p['Ingresos (miles de pesos)'])
p['Año'] = pd.to_numeric(p['Año'])
```

Creo una columna con los valores del costo por acceso por cada trimestre y anio. Seguido lo grafico en un plano

```python
#Divido los ingresos totales por el total de mbps
p['Ingreso x Acceso'] = round(p['Ingresos (miles de pesos)'] / p['Total'], 2)
```

Podemos observar como desde 2014 al 2022 el valor de cada Acceso aumento en un 500%

Teniendo en cuenta que desde Enero del 2017 a Noviembre del 2022 hubo una inflacion del 979,3% y el aumento de los Accesos es de 500% desde 2014. El coste de internet fue disminuyendo permitiendo mayor accesos a mas hogares, incrementado el uso del mismo y generando mayores ingresos.

```python
sns.lineplot(data=p,x='Año',y='Ingreso x Acceso')
```

Preparamos el Dataframe y lo exportamos

```python
p['Ingreso x Acceso'] = p['Ingreso x Acceso'].astype(str)
```

```python
p.to_csv('Datasets/Ingreso_Acceso.csv', index=False)
```
